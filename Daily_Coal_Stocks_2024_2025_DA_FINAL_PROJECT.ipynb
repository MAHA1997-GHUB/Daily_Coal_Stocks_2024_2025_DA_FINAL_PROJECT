  * **is_critical** – Flag indicating whether the plant is in a critical condition due to low coal stock.

### **Handling Missing Values (impute or drop)**
*  The dataset was carefully examined for missing values across all the columns. Most of the fields are complete with no missing values.
*   But **stock_days, is_critical, and remarks** have large gaps, while **plf_prcnt** has a few. These can be recalculated, imputed, or filled with placeholders so the dataset becomes consistent and ready for analysis.


**1. Re-derive stock_days:**
*    The **stock_days** column in the dataset has **79,100 missing values**, meaning all entries are absent.
*    Since this is a derived field, it can be systematically recalculated using the formula **stock_days = total_stock_tonnes ÷ daily_requirement_tonnes.**

  By applying this calculation, the missing values are replaced with accurate figures that reflect how many days the available coal stock can sustain operations, thereby ensuring the dataset is complete, consistent, and reliable for further analysis.


```python
# recalculate stock_days
df['stock_days'] = df['stock_days'].fillna(df['total_stock_tonnes']/df['daily_requirement_tonnes']).round(0)
```

**2. Fill plf_prcnt:**

* The Plant Load Factor Percentage **plf_prcnt** column have **299 missing** values.
* we can fill them using mean.

```python
# fill plt_prcnt
df['plf_prcnt']= df['plf_prcnt'].fillna(df['plf_prcnt'].mean()).round(2)
```

**3.Re-calculate is_critical:**
* The **is_critical** column has **67,526 missing** values, which means a large portion of this field is incomplete.
* Since is_critical is a derived indicator, it **can be recalculated logically rather than imputed**.
* The standard approach is to compare stock_days with req_normative_stock_days.

```python
# re-derive is_critical column directly
df['is_critical'] = df['stock_days'] < df['req_normative_stock_days']
df['is_critical'] = df['is_critical'].map({True: "Critical", False: "Not-Critical"})
```

**4. Fill remarks:**

* The **remarks** column have **64,316 missing** values. Mostly missing because remarks are only added when issues occur.
* So we can Fill it  with "No remarks" for consistency.

```python
# fill remarks
df['remarks'].fillna('No Remarks', inplace = True)
```

## **Data type conversion**
* The data types of all columns were carefully verified to ensure accuracy and consistency.

## **Standardizing Data Fields**
* Data fields were standardized by **converting numeric columns to proper numeric types and categorical columns to category type,** ensuring consistency, efficiency, and ease of analysis.

```python
# Converting Data fields into proper type
# Define fields
numeric_fields = [
    'capacity_of_power_station','daily_requirement_tonnes','daily_receipt_tonnes',
    'daily_consumption_tonnes','req_normative_stock_tonnes','req_normative_stock_days',
    'indigenous_stock_tonnes','import_stock_tonnes','total_stock_tonnes',
    'stock_days','plf_prcnt','actual_vs_normative_stock_prcnt'
]
categorical_fields = [
    'state_name','state_code','power_station_name','sector',
    'utility','mode_of_transport','is_critical','remarks'
]

# Convert in one loop - numerical fields
for col in numeric_fields:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert in one loop - categorical fields
for col in categorical_fields:
    df[col] = df[col].astype('category')
```

#### **3.3 Pivot Table + Sunburst Chart (Sector → Transport → Requirement)**


```python
import pandas as pd
import plotly.express as px

# Pivot table with observed=False to silence warning
pivot = df.pivot_table(
    values='daily_requirement_tonnes',
    index='sector',
    columns='mode_of_transport',
    aggfunc='mean',
    observed=False
)

# Flatten pivot table
pivot_flat = pivot.reset_index().melt(
    id_vars='sector',
    var_name='mode_of_transport',
    value_name='daily_requirement_tonnes'
).dropna()

# Sunburst chart
fig = px.sunburst(
    pivot_flat,
    path=['sector', 'mode_of_transport'],   # hierarchy
    values='daily_requirement_tonnes',
    color='daily_requirement_tonnes',
    color_continuous_scale='YlOrRd',
    title="Sunburst Chart: Daily Requirement by Sector → Transport Mode"
)

# Show label, value, and percentage with 2 decimals
fig.update_traces(
    textinfo="label+value+percent entry",
    texttemplate="%{label}<br>%{value:.0f} tonnes<br>%{percentEntry:.2%}"
)

# Styling
fig.update_layout(
    title_font=dict(size=20, family="Arial", color="black", weight='bold'),
    margin=dict(t=50, l=0, r=0, b=0)
)

fig.show()
```

**Interpretation-sunburst Chart**
- **Explain the Chart:**
This sunburst chart shows coal demand distribution across sector → transport mode, with each slice labeled by name, absolute value, and percentage share.
- **What it is saying:**

#### **3.5 Treemap (State → Sector → Transport)**

```python
import pandas as pd
import plotly.express as px

# Aggregate by state, sector, and transport
state_summary = df.groupby(
    ['state_name','sector','mode_of_transport']
)['daily_requirement_tonnes'].sum().reset_index()

# Remove rows with zero or NaN values
state_summary = state_summary[state_summary['daily_requirement_tonnes'] > 0].dropna()

# Treemap chart
fig = px.treemap(
    state_summary,
    path=['state_name','sector','mode_of_transport'],   # hierarchy
    values='daily_requirement_tonnes',
    color='daily_requirement_tonnes',
    color_continuous_scale='Plasma',
    title="Treemap: Coal Requirement by State → Sector → Transport"
)

fig.update_layout(
    title_font=dict(size=20, family="Arial", color="black", weight='bold'),
    margin=dict(t=50, l=25, r=25, b=25)
)

fig.show()
```

**Interpretation - Tree Map**
- **Explain the Chart:**
Each rectangle shows a state’s coal requirement, subdivided into sectors and then transport modes.
- **What it is saying:**
You can instantly see which states dominate demand, and within them, which sectors and transport modes are most critical. For example, Maharashtra → State Sector → Rail might occupy a large block, signaling heavy reliance on rail.
- **What features you used:**

  New variable: state_name.

  Hierarchical treemap (state → sector → transport).

  Color encoding by demand magnitude.

A state‑wise hierarchical breakdown of coal demand, making regional dependencies and transport reliance clear at a glance.
